```python
#read in dataframe
pepe = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/rare-pepes/pepecash_prices.csv',)
```

```python
pepe.head()
```

```python
#checking format of timestamp. Read.txt states datetime is in unix code, do it must be converted
pepe['Timestamp'][0].dtype
```

```python
#convert unix timestamp to date time

pepe['Timestamp'] = pd.to_datetime(pepe['Timestamp'], unit='s',)
pepe['month'] = pepe['Timestamp'].dt.month
pepe
```

```python
 #sorting by date time and resetting the index so values display in correct order 
 pepe_sorted = pepe.sort_values(by=['Timestamp'],  ascending=True)
 pepe_sorted=pepe_sorted.reset_index()
```

```python
pepe_sorted['Price'].plot()
```

```python
#plot data

plt.style.use('fivethirtyeight')


fig, ax = plt. subplots()

fig.patch.set(facecolor="white")

pepe_sorted['Price'].plot(figsize=(15,5), color="#51b253", linewidth=2.5)
ax.set(facecolor="white")

ax.set_title("The changing value of Pepe Cash", 
             fontsize=18, 
             x=.5, 
             y=1.1,
             fontweight='bold',
             fontname="Arial",
             
             )
ax.text(s="Price of Pepe Cash in U.S. Dollars", fontsize=14, x=190, y=.158);

#fix grid tick marks
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.tick_params(axis="x", direction="out", length=12, width=1, color="lightgrey")
ax.tick_params(axis="y", direction="out", length=6, width=1, color="lightgrey")
ax.set_xticklabels(['', "JAN. '17", 'MAY','SEPT.',"JAN. '18","MAR. 5, '18" ],color = "grey", x=-0.1)
ax.set_yticklabels(['0.00','0.05','0.10','$0.15'],color = "grey")
ax.set_yticks([0,0.05,0.1,0.15]);
```

```python
# More Work Here
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

candy = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/candy-power-ranking/candy-data.csv')         
```

```python
candy
```

```python
candy_win = candy.sort_values('winpercent', ascending=False).set_index('competitorname')
candy_70 = candy_win[candy_win['winpercent']>60.00]
candy_70[['sugarpercent', 'pricepercent']] = candy_70[['sugarpercent', 'pricepercent']]*100
candy_70['likyness'] = (candy_70['sugarpercent'] + candy_70['pricepercent'] + candy_70['winpercent'])/3
candy_70 = candy_70.drop(index='ReeseÕs Miniatures')

  

candy_70
```

```python
candy_70['winpercent'].plot(figsize=(20,10), marker='o', linewidth=5, alpha=0.7, colormap='viridis')
candy_70['sugarpercent'].plot(figsize=(20,10),marker='o',linewidth=5, alpha=0.7, c="teal")
candy_70['pricepercent'].plot(figsize=(20,10), marker='o',linewidth=5,alpha=0.7, c="red")
candy_70['likyness'].plot(figsize=(20,10), marker='o',linewidth=15,alpha=0.5,c="orange")

candy_70.plot.scatter('winpercent','sugarpercent',figsize=(10,10))
```
